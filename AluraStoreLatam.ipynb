# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Función para encontrar y graficar los productos más y menos vendidos
def analizar_ventas_productos(df, nombre_tienda):
    conteo_productos = df['Producto'].value_counts()
    productos_mas_vendidos = conteo_productos.head(5)  # Top 5
    productos_menos_vendidos = conteo_productos.tail(5) # Bottom 5

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    sns.barplot(x=productos_mas_vendidos.index, y=productos_mas_vendidos.values, palette="viridis")
    plt.title(f'Productos Más Vendidos - {nombre_tienda}')
    plt.xlabel('Producto')
    plt.ylabel('Cantidad Vendida')
    plt.xticks(rotation=45, ha='right')

    plt.subplot(1, 2, 2)
    sns.barplot(x=productos_menos_vendidos.index, y=productos_menos_vendidos.values, palette="magma")
    plt.title(f'Productos Menos Vendidos - {nombre_tienda}')
    plt.xlabel('Producto')
    plt.ylabel('Cantidad Vendida')
    plt.xticks(rotation=45, ha='right')

    plt.tight_layout()
    plt.show()

# Analizar las ventas de productos para cada tienda
analizar_ventas_productos(tienda, 'Tienda 1')
analizar_ventas_productos(tienda2, 'Tienda 2')
analizar_ventas_productos(tienda3, 'Tienda 3')
analizar_ventas_productos(tienda4, 'Tienda 4')
```

Gráfico 2: Distribución de las 5 Categorías de Productos Más Vendidas por Tienda (Gráfico de Barras Agrupadas)


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def obtener_top_n_categorias(df, n=5):
    return df['Categoría del Producto'].value_counts().nlargest(n)

top_categorias_tienda1 = obtener_top_n_categorias(tienda)
top_categorias_tienda2 = obtener_top_n_categorias(tienda2)
top_categorias_tienda3 = obtener_top_n_categorias(tienda3)
top_categorias_tienda4 = obtener_top_n_categorias(tienda4)

df_categorias = pd.DataFrame({
    'Tienda 1': top_categorias_tienda1,
    'Tienda 2': top_categorias_tienda2,
    'Tienda 3': top_categorias_tienda3,
    'Tienda 4': top_categorias_tienda4
}).fillna(0) # Fill NaN with 0 for plotting

# Usa reset_index() para convertir el índice en una columna llamada 'Categoría'
# y luego usa 'Categoría' como id_vars en melt()
df_categorias_melted = df_categorias.reset_index(names=['Categoría']).melt(id_vars='Categoría', var_name='Tienda', value_name='Cantidad Vendida')

# La línea de rename ya no es necesaria
# df_categorias_melted.rename(columns={'index': 'Categoría'}, inplace=True)

plt.figure(figsize=(12, 8))
sns.barplot(x='Categoría', y='Cantidad Vendida', hue='Tienda', data=df_categorias_melted)
plt.title('Top 5 Categorías de Productos Más Vendidas por Tienda')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tienda')
plt.tight_layout()
plt.show()
```
